## Data Cleaning: Dropping Irrelevant Columns

Columns that are deemed irrelevant for the analysis or future modeling (e.g., 'Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size') are dropped to streamline the dataset and focus on more impactful features.

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

## Data Cleaning: Renaming Columns

For improved readability and ease of use, several column names are renamed to more concise and intuitive labels.

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

## Data Cleaning: Handling Duplicate Rows
Duplicate entries in the dataset can skew analysis results. This step identifies and removes duplicate rows to ensure data integrity.

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

## Counting Rows Before Duplicate Removal

To quantify the impact of duplicate removal, the number of rows is counted before the duplicates are dropped.

```python
df.count()      # Used to count the number of rows
```

Approximately 989 duplicate rows were identified and removed, reducing the dataset size from 11914 to 10925 entries.

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

## Data Cleaning: Handling Missing Values

Missing values are identified using `isnull().sum()`. For this dataset, as the number of missing values in 'HP' and 'Cylinders' is relatively small compared to the total dataset size (less than 1%), these rows are dropped to maintain data quality.

```python
print(df.isnull().sum())
```

The count of non-null values for 'HP' and 'Cylinders' before dropping was 10856 and 10895 respectively out of 10925 rows, indicating the presence of missing values.

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

After dropping rows with missing values, the dataset now contains 10827 complete entries, ensuring all remaining data is valid for analysis.

```python
print(df.isnull().sum())   # After dropping the values
```

## Exploratory Data Analysis (EDA): Detecting Outliers

Outliers, data points significantly different from others, can affect statistical models. Box plots are used to visually identify outliers in key numerical features such as 'Price', 'HP', and 'Cylinders'.

```python
sns.boxplot(y=df['Price'])
```

```python
sns.boxplot(y=df['HP'])
```

## Exploratory Data Analysis (EDA): Histograms

Histograms are used to visualize the distribution of categorical data, specifically to show the frequency of cars by 'Make'. This helps in understanding the market share or representation of different car manufacturers in the dataset.

```python
df.Make.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

## Exploratory Data Analysis (EDA): Count Plots and Pie Chart

Count plots provide a visualization of the frequency of categorical variables. Here, `Make`, `Transmission`, and `Drive Mode` distributions are explored to understand the prevalence of different car attributes. A pie chart is used for `Drive Mode` to show proportions.

```python
ax = sns.countplot(x=df["Make"])
ax.tick_params(axis='x', rotation=90)
```

```python
ax = sns.countplot(x=df["Transmission"])
```

```python
trans_counts = df["Drive Mode"].value_counts()
trans_counts
```

```python
plt.pie(trans_counts, labels=trans_counts.index, autopct='%1.0f%%')
plt.show()
```

## Exploratory Data Analysis (EDA): Heat Maps for Correlation Analysis

Heat maps are effective tools for visualizing the correlation matrix between numerical features. This helps in identifying strongly correlated variables, which is crucial for understanding dependencies and for feature selection in modeling. It reveals that `Price` is strongly correlated with `HP` and `Cylinders`.

```python
plt.figure(figsize=(10,5))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])
c = numeric_df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
display(c)
```

## Exploratory Data Analysis (EDA): Scatter Plots

Scatter plots are used to visualize the relationship between two numerical variables. A scatter plot of 'HP' versus 'Price' helps to observe if there's a linear or non-linear trend, indicating how horsepower influences car price.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

## Exploratory Data Analysis (EDA): Feature Distributions

Further exploration into the distribution of individual features like 'HP' is done using `distplot` (from seaborn) and `hist` (from matplotlib) to assess normality and spread. `pairplot` provides a matrix of scatter plots for pairwise relationships and histograms/KDE for univariate distributions of selected features. Box plots and violin plots visualize the relationship between categorical ('Cylinders') and numerical ('HP', 'Price') features, highlighting value distributions and potential outliers across categories.

```python
import scipy.stats as st
```

```python
y = df['HP']
plt.figure(1); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
```

```python
plt.hist(df['HP'],orientation = 'vertical',histtype = 'bar', color ='blue')
plt.show()
```

```python
sns.pairplot(df,size = 2 ,kind ='scatter',diag_kind='kde')
plt.show()
```

```python
sns.boxplot(x="Cylinders", y="HP", data=df)
```

```python
sns.boxplot(x="Cylinders", y="Price", data=df)
```
